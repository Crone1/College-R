```r
# create random x and y co-ordinates and plot them
x = rnorm(1000)
y = rnorm(1000)
plot(x,y)
```

```r
# combine these x and y co-ordinates into 2 seperate columns
z = cbind(x,y)
head(z)
```

```r
# get the dimension of this matrix
dim(z)
```

```r
# get the variance of this table
c1 = var(z)
c1
```

```r
# get the eigenvalues of this matrix
e1 = eigen(c1)
e1
```

```r
# create new variables based off these original variables and plot them
x1 = 2*x + y
y1 = 2*x-y
plot(x1,y1)
```

```r
# combine these x and y co-ordinates into 2 seperate columns of a table
z1 = cbind(x1,y1)
head(z1)
```

```r
# get the variance of this table
c2 = var(z1)
c2
```

```r
# get the eigenvalues of this variance
e2 = eigen(c2)
e2
```

```r
# draw straight lines on the plot from these points
abline(0,1)
abline(0,-1)
```

```r
# assign the different eigenvectors to variables
u1 = e2$vectors[,1]
u1
```

```r
# assign the different evectors to variables
u2 = e2$vectors[,2]
u2
```

```r
# multiply these vectors by their transpose
t(u1) %*% u1
t(u1) %*% u2
```

```r
# create a new matrix from multiplying the eigenvectors
p = z1 %*% e2$vectors
head(p)
```

```r
# plot these points
plot(p[,1],p[,2])
```

```r
# plot these points with an 'x' and 'y' axis
plot(p[,1], p[,2], xlim= c(-10,10), ylim = c(-10, 10))
```
