# COMBINING AND SET UP

```r
# combining the data into a matrix
d = cbind(mat_d0,mat_d1,mat_d2,mat_d3,mat_d4,mat_d5,mat_d6,mat_d7,mat_d8,mat_d9)
dim(d)
```

```r
# get the variance of this combined matrix
c = var(d)
dim(c)
```

```r
# transpose it
d = t(d)
dim(d)
```

```r
# get the variance of this transposed matrix
c = var(d)
dim(c)
```

```r
# get the eigenvalues and plot them
e = eigen(c)
plot(e$values)
```

```r
# create a new matrix from these eigenvalues
p = d %*% e$vectors[, c(1:3)]
dim(p)
```

# ROTATING POINT CLOUD

```r
# set rotating angle and rotating matrix
theta = 2* pi/100
rotation = matrix(c(cos(theta), sin(theta), -sin(theta), cos(theta)), 2,2)
```

### Plot unordered rotating points

```r
# set each number as a different "p"
p1 = t(p[1:1000,])
p2 = t(p[1001:2000,])
p3 = t(p[2001:3000,])
p4 = t(p[3001:4000,])
p5 = t(p[4001:5000,])
p6 = t(p[5001:6000,])
p7 = t(p[6001:7000,])
p8 = t(p[7001:8000,])
p9 = t(p[8001:9000,])
p10 = t(p[9001:10000,])
```

```r
# rotate these points "p"
for (i in c(1:1000))
{
    plot.new()
    plot.window(xlim=c(-1600,1600), ylim = c(-900,1500))
    axis(2)
    axis(1)

    p1[c(1,3),]= rotation %*% p1[c(1,3),]
    p2[c(1,3),]= rotation %*% p2[c(1,3),]
    p3[c(1,3),]= rotation %*% p3[c(1,3),]
    p4[c(1,3),]= rotation %*% p4[c(1,3),]
    p5[c(1,3),]= rotation %*% p5[c(1,3),]
    p6[c(1,3),]= rotation %*% p6[c(1,3),]
    p7[c(1,3),]= rotation %*% p7[c(1,3),]
    p8[c(1,3),]= rotation %*% p8[c(1,3),]
    p9[c(1,3),]= rotation %*% p9[c(1,3),]
    p10[c(1,3),] = rotation %*% p10[c(1,3),]
    points(p1[1,],p1[2,], col = "orange")
    points(p2[1,],p2[2,], col = "blue")
    points(p3[1,],p3[2,], col = "red")
    points(p4[1,],p4[2,], col = "magenta")
    points(p5[1,],p5[2,], col = "brown")
    points(p6[1,],p6[2,], col = "green")
    points(p7[1,],p7[2,], col = "purple")
    points(p8[1,],p8[2,], col = "yellow")
    points(p9[1,],p9[2,], col = "pink")
    points(p10[1,],p10[2,], col = "cyan")
}
```
