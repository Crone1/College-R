```r
# read in data of the 1's with good fits
final = scan("final")
```

```r
# create matrix of data
final = matrix(final,12,874)
```

```r
# get a subset of this matrix
results2 =  final[2:11,]
results2
```

```r
# get the variance of the subset
resvar = apply(results2,1,var)
resvar
```

```r
# divide the data by its standard deviation
results3 = results2/sqrt(resvar)
results3
```

```r
# get the variance of the transpose of this
cr=var(t(results3))
cr
```

```r
# get the eigenvalues + eigenvectors
er=eigen(cr)
er
```

```r
#plot the e-values
plot(er$values)
```

```r
# multiply the e-vectors by the data
p=t(results3[1:10,]) %*% er$vectors[,1:10]
p
```

```r
# get the mean of this matrix
mp=apply(p,2,mean)
mp
```

```r
# times these means by the e-vectors
mr3=mp %*% t(er$vectors[,1:10])
mr3
```

```r
# times this matrix by the standard deviations
mr2=mr3*sqrt(resvar)
mr2
```

```r
# Create a stroke function
stroke = function(x1,y1,theta,len,width)
{
    x=c(1:16)
    y=c(1:16)
    x=x-x1
    y=y-y1
    u1=x*cos(theta)
    u2=y*sin(theta)
    #u=ones %*% t(u1) + t(ones %*% t(u2))
    u=matrix(u1,16,16)+t(matrix(u2,16,16))
    v1=-x*sin(theta)
    v2=y*cos(theta)
    #v=(ones %*% t(v1) + t(ones %*% t(v2)))
    v=matrix(v1,16,16)+t(matrix(v2,16,16))
    m = (v > 0 & v < len)
    exp(-u*u/width)*m
}
```
